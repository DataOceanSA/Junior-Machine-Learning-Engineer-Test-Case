```python
import ydata_profiling
import pandas as pd

# Read the CSV file
data = pd.read_csv("Data.csv")

# Generate the profile report
report = data.profile_report()

# Save the report to HTML or other formats
report.to_file("Data.html")
```

```python
# Generating a copy of data called df
df = data.copy()
```

```python
# Convert the Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

# Extract year, month and day
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day

# Print the modified DataFrame
print(df)
```

```python
#Replace the "," by "." in Price and Total then convert them to float
df['Price'] = df['Price'].str.replace(',', '.').astype(float)
df['Total'] = df['Total'].str.replace(',', '.').astype(float)
print(df)
```

```python
#Correct the product names
df['Product'] = df['Product'].str.replace('Bta', 'Beta')
df['Product'] = df['Product'].str.replace('Alppha', 'Alpha')
df['Product'] = df['Product'].str.replace('Dleta', 'Delta')
df['Product'] = df['Product'].str.replace('Epsilno', 'Epsilon')
df['Product'] = df['Product'].str.replace('Gamam', 'Gamma')
df['Product'] = df['Product'].str.replace('Widjet', 'Widget')

# Print the modified DataFrame
print(df)
```

```python
#Inspect the missing values
df.isnull().sum()
```

```python
from matplotlib import pyplot as plt

# Daily total sales
daily_sales = df.groupby("Day")["Total"].sum().reset_index()

# Plot daily sales time series
plt.figure(figsize=(10, 6))
plt.plot(daily_sales["Day"], daily_sales["Total"])
plt.xlabel("Days")
plt.ylabel("Total Sales")
plt.title("Daily Sales Time Series")
plt.show()
```

```python
#Total spending "Total" per customer "Costumer ID"

#Change the customer id to enhance the vizualisaion
df['Customer ID'] = df['Customer ID'].str.replace('CUST', 'C')
# Group data by customer ID and calculate total spending
customer_spending = df.groupby("Customer ID")["Total"].sum().reset_index()

# Sort data by customer ID
customer_spending.sort_values(by="Customer ID", inplace=True)

# Plot time series of total spending per customer
plt.figure(figsize=(18, 16))
plt.plot(customer_spending["Customer ID"], customer_spending["Total"])
plt.xlabel("Customer ID")
plt.ylabel("Total Spending")
plt.title("Total Spending per Customer")
plt.show()
```

```python
#Daily sales by product category

# Group data by product and day
product_daily_sales = df.groupby(["Product", "Day"])["Total"].sum().reset_index()

# Reshape data for better visualization
product_daily_sales_pivot = product_daily_sales.pivot(index="Day", columns="Product", values="Total")

# Plot daily sales by product category
plt.figure(figsize=(10, 6))
product_daily_sales_pivot.plot(kind="area", stacked=False)
plt.xlabel("Days")
plt.ylabel("Total Sales")
plt.title("Daily Sales by Product Category")
plt.legend(title="Product")
plt.show()
```

```python
#Group by product category and analyze trends using the bar chart

# Group data by product category and calculate total sales
product_category_sales = df.groupby("Product")["Total"].sum().reset_index()

# Sort data by total sales
product_category_sales.sort_values(by="Total", inplace=True)

# Define colors for the bars
colors = ["#bc4749", "#03045E", "#0077B6", "#00B4D8"]

# Create a bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(product_category_sales["Product"], product_category_sales["Total"], color=colors)
plt.xlabel("Product Category")
plt.ylabel("Total Sales")
plt.title("Total Sales by Product Category")

# Add average sales per product category as text labels
for bar, category in zip(bars, product_category_sales["Product"]):
  average_sales = df[df["Product"] == category]["Total"].mean()
  plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f"{average_sales:.2f}", ha="center", va="bottom")

plt.show()
```

```python
#Clustering daily sales
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(daily_sales[["Total"]])
daily_sales["cluster"] = kmeans.labels_

# Plot daily sales time series with clusters
plt.figure(figsize=(10, 6))
for cluster in daily_sales["cluster"].unique():
  subset = daily_sales[daily_sales["cluster"] == cluster]
  plt.plot(subset["Day"], subset["Total"], label=f"Cluster {cluster + 1}")
plt.xlabel("Days")
plt.ylabel("Total Sales")
plt.title("Clusters of Daily Sales")
plt.legend()
plt.show()
```

```python
#Clustering daily sales

# Visualize clusters
plt.figure(figsize=(10, 6))
plt.scatter(daily_sales["Day"], daily_sales["Total"], c=daily_sales["cluster"], cmap="viridis")
plt.xlabel("Days")
plt.ylabel("Total Sales")
plt.title("Daily Sales Clusters")
plt.show()

# Analyze average sales for each cluster
for cluster in daily_sales["cluster"].unique():
  cluster_sales = daily_sales[daily_sales["cluster"] == cluster]["Total"]
  print(f"Cluster {cluster}: Average Sales = {cluster_sales.mean():.2f}")
```

```python
from pyod.models.abod import ABOD

model = ABOD(contamination=0.05)  # Adjust contamination rate as needed
model.fit(daily_sales[["Total"]])
is_anomalous = model.predict(daily_sales[["Total"]])
daily_sales["is_anomalous"] = is_anomalous

# Analyze and visualize anomalies
anomalous_days = daily_sales[daily_sales["is_anomalous"] == 1]

print("Daily sales with anomaly detection:")
#print(daily_sales)
print(daily_sales.head())
```

```python
#Analyze and visualize anomalies

# Plot daily sales time series with anomaly detection
plt.figure(figsize=(10, 6))
plt.plot(daily_sales["Day"], daily_sales["Total"], label="Total Sales")
plt.scatter(anomalous_days["Day"], anomalous_days["Total"], color="red", label="Anomalies")
plt.xlabel("Days")
plt.ylabel("Total Sales")
plt.title("Daily Sales with Anomaly Detection")
plt.legend()
plt.show()

# Analyze average sales on anomalous days
normal_days = daily_sales[daily_sales["is_anomalous"] == 0]
anomalous_sales = daily_sales[daily_sales["is_anomalous"] == 1]["Total"]
num_anomalies = len(anomalous_days)
print(f"Average sales on normal days: {normal_days['Total'].mean():.2f}")
print(f"Average sales on anomalous days: {anomalous_sales.mean():.2f}")
print(f"Number of anomalies detected: {num_anomalies}")

# Analyze distribution of sales on anomalous days
plt.figure(figsize=(10, 6))
plt.hist(anomalous_sales, bins=10, edgecolor="black")
plt.xlabel("Total Sales")
plt.ylabel("Frequency")
plt.title("Distribution of Sales on Anomalous Days")
plt.show()
```
